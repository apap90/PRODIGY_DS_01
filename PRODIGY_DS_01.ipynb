## Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

Sample Dataset :- https://data.worldbank.org/indicator/SP.POP.TOTL

### Importing libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Loading the dataset:

```python
df = pd.read_csv('/content/API_SP.POP.TOTL_DS2_en_csv_v2_339468.csv')
df.head() # Displaying the first few rows of the dataset
```

### Filtering and cleaning the data:

```python
df = df[['Country Name', '2022']]  # Selecting only the relevant columns (Country Name and 2022 population)
```

```python
df = df.dropna()  # Dropping rows with missing values (if any)
```

### Sort by population and select top 10 countries:

```python
df = df.sort_values(by='2022', ascending=False).head(10)
```

### Plotting the bar chart:

```python
plt.figure(figsize=(10, 6))
plt.barh(df['Country Name'], df['2022'], color='skyblue')
plt.xlabel('Population')
plt.title('Top 10 Countries by Population in 2019')
plt.gca().invert_yaxis()  # Invert y-axis to have the largest bar at the top
plt.tight_layout()
plt.show()
```
